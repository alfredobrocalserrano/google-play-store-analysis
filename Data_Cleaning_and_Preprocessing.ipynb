```python
import pandas as pd

df = pd.read_csv('Google-Playstore.csv')

print("Forma:", df.shape)
print("\nColumnas:")
print(df.dtypes)
print("\nValores nulos:")
print(df.isnull().sum())
print("\nMuestra:")
df.head(3)
```

```python
# 1. Quedarnos solo con columnas útiles
cols = ['App Name', 'Category', 'Rating', 'Rating Count',
        'Installs', 'Free', 'Price', 'Size', 'Content Rating']
df = df[cols].copy()

# 2. Eliminar nulos en columnas clave
print(f"Filas antes: {len(df)}")
df = df.dropna(subset=['Rating', 'Category', 'Installs'])
print(f"Filas tras eliminar nulos: {len(df)}")

# 3. Limpiar columna Installs (quitar + y comas)
df['Installs'] = df['Installs'].str.replace('+', '', regex=False)
df['Installs'] = df['Installs'].str.replace(',', '', regex=False)
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')

# 4. Limpiar columna Price
df['Price'] = pd.to_numeric(df['Price'], errors='coerce').fillna(0)

# 5. Eliminar filas con valores extraños en Rating
df = df[df['Rating'] <= 5]
df = df[df['Rating'] > 0]

print(f"Filas finales: {len(df)}")
print(f"\nDistribución Free vs Paid:")
print(df['Free'].value_counts())
print(f"\nTop 5 categorías:")
print(df['Category'].value_counts().head())
```

```python
df.to_csv('playstore_limpio.csv', index=False)
print(f"Exportado correctamente — {len(df)} filas")

from google.colab import files
files.download('playstore_limpio.csv')
```
